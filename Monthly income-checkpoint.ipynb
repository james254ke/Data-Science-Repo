<h3 style="color:green;">Create Wrangle Function</h3>

```python
def wrangle(filepath):
    #Read the data from the directory as csv
    data=pd.read_csv(filepath).set_index("employeenumber")    
    
    #Extract the important features needed for the study
    data_New=data[["monthlyincome","age","education","gender","overtime","totalworkingyears"]]
    data=data_New
    
    
    return data
```

<h2 style="text-decoration:underline;color:skyblue;"><b>3.Split the Data</b></h2>

**Assign the Explanatory Variables X and the target variable assign it y**

```python
target="monthlyincome"
X=data.drop(columns=target)
y=data[target]
```

<li style="color:green;"><font size="+2"><b>Split the Data into train and test</b></font></li>

**Test data will have a split of 20% of the total data while the training data will be 80%**

```python
X_train,X_test,y_train,y_test=train_test_split(
    X,y,test_size=0.20, random_state=42
)
print("X_train shape:",X_train.shape)
print("y_train shape:",y_train.shape)
print("X_test shape:",X_test.shape)
print("y_test shape:",y_test.shape)
```

<h2 style="text-decoration:underline;color:skyblue;"><b>1.Baseline</b></h2>

<li style="color:green;"><font size="+1"><b>Get the mean_absolute_error which is important in evaluating the perfomance of the model</b></font></li>

```python
y_mean=y_train.mean()
y_pred_baseline=[y_mean]*len(y_train)
print("Mean Monthly Income:",round(y_mean,2))
print("Baseline Accuracy:",mean_absolute_error(y_train,y_pred_baseline))
```

<h2 style="text-decoration:underline;color:skyblue;"><b>2.Iterate</b></h2>

**Instantiate Linear Regression**

```python
model=LinearRegression()
```

**Fit the data in the model**

```python
model.fit(X_train,y_train)
LinearRegression()
```

<h2 style="text-decoration:underline;color:skyblue;"><b>3.Evaluate the Model</b></h2>

```python
y_pred_training=model.predict(X_train)
print("Training MAE:",mean_absolute_error(y_train,y_pred_training))
```

**Our predicted Model has perfomed well as it beat the baseline by over 1000£**

**use the test data to evaluate the perfomance of the model**

```python
y_pred_test=model.predict(X_test)
print("Training MAE:",mean_absolute_error(y_test,y_pred_test))
```

**Our predicted test model also perfoms well it beat the baseline**

**Create a Function that predicts the Monthly income of employees**

```python
def income_pred(age,education,gender,overtime,totalworkingyears):
    data={
        "age":age,
        "education":education,
        "gender":gender,
        "overtime":overtime,
        "totalworkingyears":totalworkingyears
    }
    
    data1=pd.DataFrame(data,index=[0])
    prediction=model.predict(data1).round(2)[0]
    return f"Predicted monthly income:£{prediction}"
```

```python
#prediction cell ,use this cell to input the values so as to get the monthly income
#for instance a male(1) aged 45 with an education of more than 5 years(after highschool,has no overtime(0), and has experience
#of 9 years earns 4810 pounds)
income_pred(45,5,1,0,9)
```
